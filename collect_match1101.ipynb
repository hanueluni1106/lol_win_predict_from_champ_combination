```python
import requests

API_KEY = ''
```

#### 소환사 아이디 수집

##### summonerid => puuid

```python
def transform_sid_to_puuid(summonerid):
    puuid = requests.get(f"https://kr.api.riotgames.com/lol/summoner/v4/summoners/{summonerid}?api_key={API_KEY}").json()["puuid"]
    return puuid


def get_ids_challenger_summoner():

    summoner_ids = []
    entries = requests.get(f"https://kr.api.riotgames.com/lol/league/v4/challengerleagues/by-queue/RANKED_SOLO_5x5?api_key={API_KEY}").json()["entries"]  #300명 챌린저 유저 정보 담김
    for entry in entries:
        summoner_ids.append(entry["summonerId"])
    
    return summoner_ids


def get_ids_grandmaster_summoner():

    summoner_ids = []
    entries = requests.get(f"https://kr.api.riotgames.com/lol/league/v4/grandmasterleagues/by-queue/RANKED_SOLO_5x5?api_key={API_KEY}").json()["entries"]  #300명 챌린저 유저 정보 담김
    for entry in entries:
        summoner_ids.append(entry["summonerId"])
    
    return summoner_ids

def get_ids_master_summoner():

    summoner_ids = []
    entries = requests.get(f"https://kr.api.riotgames.com/lol/league/v4/masterleagues/by-queue/RANKED_SOLO_5x5?api_key={API_KEY}").json()["entries"]  #300명 챌린저 유저 정보 담김
    for entry in entries:
        summoner_ids.append(entry["summonerId"])
    
    return summoner_ids
```

#### 소환사 경기 아이디 수집

```python
def get_ids_match(puuid):
    matchids = requests.get(f"https://asia.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?queue=420&type=ranked&start=0&count=60&api_key={API_KEY}").json()
    return matchids
```

#### 경기 챔피언 조합 수집

```python
def get_champcomb_match(matchid):

    label = None
    data = []

    match_data = requests.get(f"https://asia.api.riotgames.com/lol/match/v5/matches/{matchid}?api_key={API_KEY}").json()["info"]
    
    #라벨 (승패 1 0)
    label = 1 if match_data["teams"][0]["win"] == True else 0

    #챔프 조합
    users = match_data["participants"]  #소환사 10명 
    for user in users:
        data.append(user["championName"])

    return data,label
```

```python
get_champcomb_match('KR_6196182921')
```

#### 챌린저 한정 실험

```python
summonerids = get_ids_challenger_summoner()
puuids = [transform_sid_to_puuid(id) for id in summonerids]
len(puuids)
```

```python
matchids = []
for puuid in puuids:
    matchids.extend(get_ids_match(puuid))

len(matchids)
```

```python
matchids = list(set(matchids))
len(matchids)
```

```python
for matchid in matchids:
    data,label = get_champcomb_match(matchid)
    print(len(data),label)
```
